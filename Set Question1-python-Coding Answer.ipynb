```python
#Write a Python program that prints numbers from 1 to 5.
for i in range(1,6):
    print(i,end=" ")
```

```python
#.Write a Python program to print the first 5 even numbers.
for i in range(1,11):
    if i%2==0:
        print(i,end=" ")
```

```python
#Write a Python program to print the sum of numbers from 1 to 10.
total=0
for i in range(1,11):
    total= total + i
print(total)
```

```python
#Write a Python function to find the factorial of 5.
fact=1
for i in range(1,6):
    fact=fact*i
print(fact)    
```

```python
#Write a Python program to check if a number is prime.
num=int(input())
if (num%2==0):
    print(num,"is not prime number")
else:
    print(num,"is prime number")
```
