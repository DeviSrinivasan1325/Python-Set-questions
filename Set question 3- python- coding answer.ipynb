```python
#3.Find Even and Odd Numbers in a List 
#Question: Write a Python program to separate even and odd numbers from the list [12, 17, 19, 24, 35, 40, 56].
lists= [12, 17, 19, 24, 35, 40, 56]
Even_number=[]
Odd_number=[]
for i in lists:
    if(i%2==0):
        Even_number.append(i)
    else:
        Odd_number.append(i)

print("Even number=", Even_number)
print("Odd number=", Odd_number)
```

```python
#4. Reverse a Number
#Question: Write a Python program to reverse the number 12345.
num="12345"
Rev_num=num[::-1]
print(Rev_num)
```

```python
#5. Find the Second Largest Number in a List
#Question: Write a Python program to find the second largest number in the list [12, 45, 98, 34, 23].
lists=[12, 45, 98, 34, 23]
lists.sort()       #12,23,34,45,98
second_large= lists[-2]
print("The second largest number is",second_large)
```

```python
#manual loop
lists=[12, 45, 98, 34, 23]
largest_num=max(lists) #98
sec_num=-1
for num in lists:
    if(num!=largest_num and num>sec_num):
        sec_num=num
print("The second largest number is",sec_num)
```
