```python
#5.What is the issue with this function?
#Ans: Logical error is present in this code. if the n%1==0 return true in the first factor itself it will return true instead of checking number is prime or not.
def is_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return True
    return False
```

```python
#8.What is the issue with this function?
#Ans: initialy total=0, then only sum of numbers will be correctly print.
def sum_numbers(numbers):
    total = 0
    for num in numbers:
        total += num
    return total
```

```python
#9.What is the issue with this recursive function?
#Ans: here base case is required, where the 0!=1.
def factorial(n):
    if n==0:
        return 1
    for i in n:
        return n * factorial(n-1)
```
